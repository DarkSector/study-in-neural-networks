# Activation functions

## Softmax

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
```

### Softmax is defined as exp(val) / sum(exp(values))

```python
inputs = np.array([-2, -1, 0, 1, 2])
denominator = sum([np.exp(i) for i in inputs])
outputs = [(np.exp(val) / denominator) for val in inputs]
```

```python
fig, ax = plt.subplots()
ax.scatter(inputs, outputs)
ax.grid(True)
ax.set_xlabel("Values from -2 to 2")
ax.set_ylabel("Softmax Values")
ax.set_title("Softmax distribution")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.scatter(inputs, outputs)
plt.yscale("log")
ax.set_xlabel("Values from -2 to 2")
ax.set_ylabel("Softmax Values")
ax.set_title("Softmax distribution in Log Scale $\sum\sigma$ = %g" % denominator)
plt.show()
```

### Softmax through PyTorch

```python
# create instance of softmax activation class
softfun = nn.Softmax(dim=0)
sigmaT = softfun(torch.Tensor([1, 2, 3]))
print(sigmaT)
```

## Logarithm

```python
# 200 linearly spaced x values
inputs = np.linspace(.0001, 1, 200)
logx = np.log(inputs)
expx = np.exp(inputs)

fig, ax = plt.subplots()
ax.scatter(inputs, logx)
ax.scatter(inputs, expx)
ax.set_xlabel("Values from .0001 to 1")
ax.set_ylabel("Log values")
ax.set_title("Log distribution")
plt.show()
```
